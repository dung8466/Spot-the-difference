```python
import cv2
import imutils
import numpy as np
from skimage import metrics
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('images/camels1.jpg')
img1 = cv2.resize(img1, (600,360))
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img_height = img1.shape[0]
kernel = np.ones((5,5),np.uint8)
```

```python
img0 = img1
low_green = np.array([25, 52, 72])
high_green = np.array([102, 255, 255])
# convert BGR to HSV
imgHSV = cv2.cvtColor(img0, cv2.COLOR_BGR2HSV)
# create the Mask
mask = cv2.inRange(imgHSV, low_green, high_green)
# inverse mask
mask = 255-mask
res = cv2.bitwise_and(img0, img0, mask=mask)
img0 = res
gray0 = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)
cv2.imwrite('img1.png', img0)
diff0 = cv2.absdiff(gray1, gray0)
cv2.imshow("diff(img1, img0)", diff0)
thresh0 = cv2.threshold(diff0, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
cv2.imshow("Threshold", thresh0)
dilate0 = cv2.dilate(thresh0, kernel, iterations=2) 
cv2.imshow("Dilate", dilate0)
contours0 = cv2.findContours(dilate0.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contours0 = imutils.grab_contours(contours0)
x = np.zeros((img_height,10,3), np.uint8)
lv0 = np.hstack((img1, x, img0))
cv2.imshow("Lv0", lv0)
for contour in contours0:
    if cv2.contourArea(contour) > 100:
        x, y, w, h = cv2.boundingRect(contour)
        #cv2.rectangle(img1, (x, y), (x+w, y+h), (0,0,255), 2)
        cv2.rectangle(img0, (x, y), (x+w, y+h), (0,0,255), 2)
x = np.zeros((img_height,10,3), np.uint8)
result0 = np.hstack((img1, x, img0))


img2 = cv2.medianBlur(img1, 5)
erosion = cv2.erode(img2, kernel, iterations=1)
output = cv2.dilate(erosion, kernel, iterations=1)
gray2 = cv2.cvtColor(output, cv2.COLOR_BGR2GRAY)
cv2.imwrite('img2.png', output)
diff1 = cv2.absdiff(gray1, gray2)
cv2.imshow("diff(img1, output)", diff1)
thresh1 = cv2.threshold(diff1, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
cv2.imshow("Threshold", thresh1)
dilate1 = cv2.dilate(thresh1, kernel, iterations=2) 
cv2.imshow("Dilate", dilate1)
contours1 = cv2.findContours(dilate1.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contours1 = imutils.grab_contours(contours1)
x = np.zeros((img_height,10,3), np.uint8)
lv1 = np.hstack((img1, x, img2))
cv2.imshow("Lv1", lv1)
for contour in contours1:
    if cv2.contourArea(contour) > 100:
        x, y, w, h = cv2.boundingRect(contour)
        #cv2.rectangle(img1, (x, y), (x+w, y+h), (0,0,255), 2)
        cv2.rectangle(img2, (x, y), (x+w, y+h), (0,0,255), 2)
x = np.zeros((img_height,10,3), np.uint8)
result1 = np.hstack((img1, x, img2))

img3 = img1
img_hsv=cv2.cvtColor(img3, cv2.COLOR_BGR2HSV)
# lower mask (0-10)
lower_red = np.array([50,50,0])
upper_red = np.array([255,255,50])
mask0 = cv2.inRange(img_hsv, lower_red, upper_red)
# upper mask (170-180)                                                                                                                                                        
lower_red = np.array([50,50,50])    
upper_red = np.array([255,255,180])    
mask1 = cv2.inRange(img_hsv, lower_red, upper_red)    
# join my masks    
mask = mask0+mask1    

output_img = img3.copy()    
output_img[np.where(mask!=0)] = [125, 50, 50]
cv2.imwrite("img3.png",output_img)
img3 = output_img
gray3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)
diff2 = cv2.absdiff(gray1, gray3)
cv2.imshow("diff(img1, img3)", diff2)
thresh2 = cv2.threshold(diff2, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
cv2.imshow("Threshold", thresh2)
dilate2 = cv2.dilate(thresh2, kernel, iterations=2) 
cv2.imshow("Dilate", dilate2)
contours2 = cv2.findContours(dilate2.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contours2 = imutils.grab_contours(contours2)
lv2 = np.hstack((img1, x, img3))
cv2.imshow("Lv2", lv2)
for contour in contours2:
    if cv2.contourArea(contour) > 100:
        x, y, w, h = cv2.boundingRect(contour)
        #cv2.rectangle(img1, (x, y), (x+w, y+h), (0,0,255), 2)
        cv2.rectangle(img3, (x, y), (x+w, y+h), (0,0,255), 2)
x = np.zeros((img_height,10,3), np.uint8)
result2 = np.hstack((img1, x, img3))

cv2.imshow("Lv0 Result", result0)
cv2.imshow("Lv1 Result", result1)
cv2.imshow("Lv2 Result", result2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
